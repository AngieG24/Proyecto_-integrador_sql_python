```python
# Valores nulos en cada DataFrame
def detectar_nulos(df, nombre):
    nulos = df.isnull().sum()
    porcentaje = (df.isnull().mean() * 100).round(2) # Convertir a porcentaje y redondear a 2 decimales
    df_nulos = pd.DataFrame({"Valores Nulos": nulos, "Porcentaje (%)": porcentaje})
    df_nulos = df_nulos[df_nulos["Valores Nulos"] > 0] # Filtrar solo las columnas que tienen valores nulos
    print(f"\nüîç Valores nulos en {nombre}:")
    display(df_nulos)
```

```python
detectar_nulos(df_city, "City")
```

#### -üìù Exploraci√≥n para validar si es posible completar los valores nulos que hay en el df de City, columna District

```python
# Filtrar filas con valores nulos en la columna District y mostrar el resto de la informaci√≥n
print("\nüîç Filas con valores nulos en la columna 'District' en City:")
df_nulos_district = df_city[df_city["District"].isnull()]
df_nulos_district
```

```python
df_country
```

Se realiz√≥ una **uni√≥n** con la tabla `country` para obtener m√°s contexto sobre cada ciudad, incluyendo el **nombre del pa√≠s**, la **regi√≥n** y el **√°rea superficial**. Esto permiti√≥ analizar mejor la informaci√≥n y facilitar la b√∫squeda de los distritos correctos.

```python
df_nulos_district = df_nulos_district.merge(
    df_country[["Code", "Name", "Region", "SurfaceArea"]].rename(columns={"Name": "NamePais"}), 
    left_on="CountryCode",  # Columna en df_nulos_district
    right_on="Code",        # Columna en df_country
    how="left"
)
df_nulos_district
```

###### üåç B√∫squeda de informaci√≥n externa
Debido a la ausencia de datos en la fuente original, consul√©s fuentes externas para determinar los distritos correspondientes a cada ciudad con valores nulos


```python
# Completar valores manualmente en el DataFrame
df_city.loc[df_city["ID"] == 3285, "District"] = "Taichung"
df_city.loc[df_city["ID"] == 3293, "District"] = "Kaohsiung"
df_city.loc[df_city["ID"] == 3294, "District"] = "Taoyuan"
df_city.loc[df_city["ID"] == 3563, "District"] = "Zulia"  # Estado en Venezuela

# Verificar cambios
df_city[df_city["ID"].isin([3285, 3293, 3294, 3563])]
```
